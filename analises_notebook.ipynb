# Conhecimento do Problema

As análises dos dados são realizadas lentamente, com dificuldade, muitas vezes trazendo dados incosistentes, produzindo erros e tomadas de decisão incorretas. 

É necessário realizar as análises abaixo de forma eficiente:
- Resultado final dos alunos se foi aprovado ou reprovado ao concluir o curso;
- Quantos alunos foram aprovados e quantos foram reprovados no curso;
- Comparação de rendimento de alunos por turma e região;

O aluno é considerado aprovado se sua média for igual ou superior a 6,5.

# Compreensão dos dados

Os dados são gerados em um arquivo base_alunos.csv através de um sistema web do curso em que os professores o alimentam diariamente.

A tabela é composta pelas seguintes colunas com suas respectivas definições:

- Matrícula: matrícula do aluno na escola;
- Nome: nome do aluno;
- Gênero: sigla que representa o gênero do aluno (M ou F);
- Descrição: descrição do gênero do aluno. Masculino para M, e Feminino para F;
- Turma: turma do aluno;
- Região: região da turma. Ceará para as turmas A e B, Pernambuco para as turmas C e D;
- Nota1: primeira nota do aluno;
- Nota2: segunda nota do aluno;
- Nota3: terceira nota do aluno;
- Nota4: quarta nota do aluno;

Cálculo da média e do resultado final dos alunos

```python
df['media']=(df['nota1']+df['nota2']+df['nota3']+df['nota4'])/4
```

```python
df
```

```python
df.loc[df['media']>=6.5,'resultado']='Aprovado'
df.loc[df['media']<6.5,'resultado']='Reprovado'
```

```python
df
```

Análise e visualização dos dados (relatório)

```python
r_total = df.groupby('resultado')['matricula'].count()
r_total = r_total.reset_index()
r_total
```

```python
mp.bar(r_total['resultado'],r_total['matricula'])
```

```python
mp.bar('aprovado',r_total['matricula'][0], label=r_total['matricula'][0],color='green')
mp.bar('reprovado',r_total['matricula'][1], label=r_total['matricula'][1],color='red')
mp.title("Distribuição de aprovação dos alunos no curso")
mp.legend()
mp.show()
```

```python
r_turma_regiao = df.groupby(['resultado','turma','regiao'])['matricula'].count()
r_turma_regiao = r_turma_regiao.reset_index()
r_turma_regiao
```

```python
colors = ["#008000", "#FF0000"]
sns.set_palette(sns.color_palette(colors))
```

```python
sns.catplot(y='regiao', x='matricula',hue='resultado', kind='bar',data=r_turma_regiao,ci=None)
```
